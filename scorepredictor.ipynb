```python
import numpy as np
import pandas as pd
```

```python
newdf = pd.read_csv("Innings_with_scores.csv")
newdf.head()
```

```python
data = newdf[['Team1','Team2','Score1','Score2','venue']]
data.head(15)
```

```python
df1 = ['KKR','RCB','CSK','MI','DD','SRH','RR','KXIP']
df2= ['KKR','RCB','CSK','MI','DD','SRH','RR','KXIP']
df3 = ['Kolkata','Bangalore','Chennai','Mumbai','Delhi','Hyderabad (Deccan)','Jaipur','Mohali']
for j in range(0,len(df1)):
    a = []
    b = []
    for k in range(0,len(df2)):
        for i in range(0,len(data)):
            if(j==k):
                break
            if(data.iloc[i][0]== df1[j] and data.iloc[i][1]==df2[k] and data.iloc[i][4]==df3[j]):
                a.append(data.iloc[i][2])
                b.append(data.iloc[i][3])
        if(j==k):
            pass
        else:
            print(df1[j],'=',round(sum(a)/len(a)))
            print(df2[k],'=',round(sum(b)/len(b)))
            print('________________')
```
